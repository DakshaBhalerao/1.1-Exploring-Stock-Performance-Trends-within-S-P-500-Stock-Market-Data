## Analyze the moving average of various stocks

```python
all_data['close']
```

```python
# all_data.rolling?
```

```python
all_data['close'].rolling(window = 10).mean().head(14)
```

```python
new_data = all_data.copy()
```

```python
new_data
```

```python
ma_day = [10,20,30]

for m in ma_day:
    new_data['close'+str(m)] = new_data['close'].rolling(window = m).mean()
```

```python
new_data.tail(10)
```

```python
#using pandas plot instead of matplotlib for this only change we need to do is we need to set date as index
```

```python
new_data.set_index('date',inplace=True)
```

```python
new_data
```

```python
new_data.columns
```

```python
plt.figure(figsize=(20,12))

for index , company in enumerate(tech_list,1):
    plt.subplot(2,2,index)
    filter2 = new_data['Name'] == company
    df2 = new_data[filter2]
    df2[['close10', 'close20', 'close30']].plot(ax = plt.gca())
    plt.title(company)
```
